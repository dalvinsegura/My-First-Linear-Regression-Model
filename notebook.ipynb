# My first model: Linear Regression
This model will learn how to convert celsius to fahrenheit

````python
import pandas as pd
datos = pd.read_csv('celsius.csv')
````

````python
datos.head()
````

````python
import seaborn as sns

sns.scatterplot(data=datos,x="celsius", y="fahrenheit", hue="fahrenheit", palette="coolwarm")
````

Let's define ```X``` "input data" and ```y``` "tags"

X and y are not proccesed to a value format that model can use it for learning, so we need to reshape it from this ```[0,1,2,3]``` to this ```[[0],[1],[2],[3]]``` with ```reshape(-1,1)```

````python
X = datos["celsius"]
y = datos["fahrenheit"]

X_processed = X.values.reshape(-1, 1)
y_processed = y.values.reshape(-1, 1)
````

After we proccessed our data let's train the model.

In this case our data are lineal (we can check it out with ```sns.scatterplot()``` that's why we going to use Linear Regression.

To train our model with use ```model.fit()``` and give it our data processed before.

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_processed, y_processed)
````

### It's show time
let's test our model trained!

````python
model.predict([[8]])
````

Let's check how effective is our model with ```score()```

````python
model.score(X_processed, y_processed)
````
